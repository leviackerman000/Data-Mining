```python
import numpy as np
import matplotlib.pyplot as plt

class NearestNeighborClassifierManual:
    def __init__(self):
        self.X_train = None
        self.y_train = None

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []
        for x_test in X_test:
            distances = np.linalg.norm(self.X_train - x_test, axis=1)
            nearest_neighbor_index = np.argmin(distances)
            nearest_neighbor_label = self.y_train[nearest_neighbor_index]
            predictions.append(nearest_neighbor_label)
        return predictions

# Generate random points for training
num_points = 100
random_x = np.random.uniform(minx, maxx, num_points)
random_y = np.random.uniform(miny, maxy, num_points)
X_train = np.column_stack((random_x, random_y))
y_train = np.random.randint(0, 2, num_points)  # Random labels for demonstration

# Create and fit KNN model
knn_model = NearestNeighborClassifierManual()
knn_model.fit(X_train, y_train)

# Generate random test points for prediction
num_test_points = 100
test_x = np.random.uniform(minx, maxx, num_test_points)
test_y = np.random.uniform(miny, maxy, num_test_points)
X_test = np.column_stack((test_x, test_y))

# Make predictions using KNN model
predictions = knn_model.predict(X_test)

# Plot clusters as circles on the map
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_facecolor('#EEEEEE')

for i in range(num_test_points):
    color = 'red' if predictions[i] == 1 else 'blue'
    circle = plt.Circle((test_x[i], test_y[i]), radius=0.1, color=color, alpha=0.5)
    ax.add_artist(circle)

# Plot other layers (buildings, land use, etc.)
buildings.plot(ax=ax, color='#FFFFFF')
landuse.plot(ax=ax, color='#28B463')
natural.plot(ax=ax, color='#F8C471')
places.plot(ax=ax, color='#99A3A4')
points.plot(ax=ax, color='#48C9B0')
railways.plot(ax=ax, color='#4A235A')
roads.plot(ax=ax, color='#000000')
waterways.plot(ax=ax, color='#85C1E9')
# Add other layers as needed

plt.show()
```
